```python
video_paths = ['GH012423.MP4','GH012424.MP4','GH012430 2.mov','GH012446.MP4','GH013300.mov','GH013302.MP4','GH013336.MP4','GH013337.MP4','GH013339.MP4','GH013375.MP4','GH013415 2.mov']

for video in video_paths:
    cam = cv2.VideoCapture(f"/Users/kate/Downloads/{video}")
    
    currentframe = 0
    
    while(True): 
        ret,frame = cam.read() 
    
        if ret: 
            if currentframe % 20 == 0:
                name = './dive_video_frames/' + video[2:6] + '_frame' + str(currentframe) + '.jpg'
                print ('Creating...' + name) 
                cv2.imwrite(name, frame) 
            else:
                pass
            currentframe += 1
            
        else: 
            break
        
    cam.release() 
    cv2.destroyAllWindows()
```
